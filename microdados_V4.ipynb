```python
def plotar_histograma_percentual(
    df,
    coluna_x,
    coluna_hue,
    dicionario_x=None,
    dicionario_hue=None,
    ordem_x=None,
    ordem_hue=None,
    titulo='',
    rotacao_xticks=0,
    rotacao_xticks_barra=0,
    tamanho_texto=12,
    paleta_cores='Set2',
    normalizar_por='x',
    lado_legenda='right'
):
    """
    Plota barras lado a lado mostrando a porcentagem, com normalização flexível.

    Parâmetro:
      - normalizar_por: 
            'x'   => Normaliza cada linha (cada categoria de coluna_x soma 100%).
            'hue' => Normaliza cada coluna (cada categoria de coluna_hue soma 100%).

    Exemplo de uso:
      Para analisar a distribuição de acesso à internet dentro de cada nível de desempenho,
      se 'desempenho' está em coluna_x e 'acesso' em coluna_hue, use normalizar_por='x'.
    """
    # Monta o crosstab com contagens absolutas
    crosstab_absoluto = pd.crosstab(
        index=df[coluna_x],
        columns=df[coluna_hue]
    )
    
    # Normalização: escolha a base de acordo com o parâmetro
    if normalizar_por == 'x':
        # Normaliza por linha: para cada valor de coluna_x, as porcentagens somam 100%
        crosstab_percentual = crosstab_absoluto.div(crosstab_absoluto.sum(axis=1), axis=0) * 100
    elif normalizar_por == 'hue':
        # Normaliza por coluna: para cada valor de coluna_hue, as porcentagens somam 100%
        crosstab_percentual = crosstab_absoluto.div(crosstab_absoluto.sum(axis=0), axis=1) * 100
    else:
        raise ValueError("O parâmetro normalizar_por deve ser 'x' ou 'hue'.")
    
    # Converte o crosstab para formato longo (necessário para o Seaborn)
    df_plot = crosstab_percentual.reset_index().melt(
        id_vars=coluna_x,
        var_name=coluna_hue,
        value_name='porcentagem'
    )
    
    # Define ordem para coluna_x, se não fornecida
    if ordem_x is None:
        if dicionario_x:
            ordem_x = list(dicionario_x.keys())
        else:
            ordem_x = sorted(crosstab_absoluto.index.unique())
    
    # Define ordem para coluna_hue, se não fornecida
    if ordem_hue is None:
        if dicionario_hue:
            ordem_hue = list(dicionario_hue.keys())
        else:
            ordem_hue = sorted(crosstab_absoluto.columns.unique())
    
    # Cria o gráfico de barras
    plt.figure(figsize=(11, 8), dpi=80)
    ax = sns.barplot(
        data=df_plot,
        x=coluna_x,
        y='porcentagem',
        hue=coluna_hue,
        palette=paleta_cores,
        order=ordem_x,
        hue_order=ordem_hue
    )
    
    # Adiciona o valor percentual no topo de cada barra
    for barra in ax.patches:
        altura = barra.get_height()
        if altura > 0.01:
            x_pos = barra.get_x() + barra.get_width() / 2.0
            ax.text(
                x_pos,
                altura + 0.25,
                f"{altura:.1f}%",
                ha="center",
                rotation=rotacao_xticks_barra,
                size=tamanho_texto
        )
    
    # Ajusta os rótulos do eixo x se um dicionário for fornecido
    if dicionario_x:
        novos_rotulos = [dicionario_x.get(cat, str(cat)) for cat in ordem_x]
        ax.set_xticklabels(novos_rotulos, fontsize=20)
    
    # Ajusta a legenda (para o hue) se um dicionário for fornecido
    if dicionario_hue:
        handles, labels = ax.get_legend_handles_labels()
        novos_labels = []
        for lbl in labels:
            try:
                lbl_int = int(lbl)
                novo = dicionario_hue.get(lbl_int, lbl)
            except ValueError:
                novo = dicionario_hue.get(lbl, lbl)
            novos_labels.append(novo)
        ax.legend(
            handles,
            novos_labels,
            title=coluna_hue,
            loc='upper ' + lado_legenda,
            borderaxespad=0,
            fontsize=15,
            title_fontsize=15
        )
    else:
        ax.legend(
            title=coluna_hue,
            loc='upper ' + lado_legenda,
            borderaxespad=0,
            fontsize=15,
            title_fontsize=15
        )
    
    plt.title(titulo, pad=15)
    plt.xlabel(coluna_x, fontsize=17)
    plt.ylabel('Porcentagem (base: {})'.format('coluna_x' if normalizar_por=='x' else 'coluna_hue'), fontsize=20)
    plt.xticks(rotation=rotacao_xticks)
    plt.tight_layout()
    plt.show()
```

```python
# Agrupando por sexo e calculando a média das notas
desempenho_sexo = microdados.groupby('TP_SEXO')[['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT']].median()
desempenho_sexo.index = desempenho_sexo.index.map({'F':'Feminino', 'M':'Masculino'})

# Renomeando as colunas para tornar a legenda mais clara
desempenho_sexo.rename(columns={
    'NU_NOTA_CN': 'Ciências da Natureza',
    'NU_NOTA_CH': 'Ciências Humanas',
    'NU_NOTA_LC': 'Linguagens e Códigos',
    'NU_NOTA_MT': 'Matemática'
}, inplace=True)

# Plotando o desempenho por sexo com valores no topo das bar17s
ax = desempenho_sexo.plot(kind='bar', figsize=(11, 8), fontsize=20)
plt.xlabel('Sexo', fontsize=17)
plt.ylabel('Nota Média', fontsize=20)
plt.xticks(rotation=0)
plt.ylim(400, None)
ax.legend(fontsize=15, title_fontsize=15)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.2f}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        xytext=(0, 5),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```
