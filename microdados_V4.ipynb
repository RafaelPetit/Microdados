```python
def plotar_histograma(
    df,
    coluna_x,
    coluna_hue,
    dicionario_x=None,
    dicionario_hue=None,
    ordem_x=None,
    ordem_hue=None,
    titulo='',
    rotacao_xticks=0,
    rotacao_xticks_barra=0,
    tamanho_texto=12,
    paleta_cores='Set2'
):
    """
    Plota um histograma de contagem (countplot) usando Seaborn, com controle de
    ordem e mapeamento de rótulos (labels) através de dicionários.

    Parâmetros
    ----------
    df : pd.DataFrame
        DataFrame contendo os dados que serão plotados.

    coluna_x : str
        Nome da coluna (no DataFrame) que será usada no eixo X.

    coluna_hue : str
        Nome da coluna (no DataFrame) que será usada como hue (categoria que
        divide as barras).

    dicionario_x : dict, opcional
        Mapeamento {valor_no_df: rótulo_desejado} para substituir os rótulos
        que aparecem no eixo X.

    dicionario_hue : dict, opcional
        Mapeamento {valor_no_df: rótulo_desejado} para substituir os rótulos
        na legenda (hue).

    ordem_x : list, opcional
        Lista que determina a ordem das categorias no eixo X.
        Se None e existir dicionario_x, a ordem assumirá as chaves de dicionario_x.
        Caso contrário, usará a ordem natural (unique()) dos dados.

    ordem_hue : list, opcional
        Lista que determina a ordem das categorias na legenda (hue).
        Se None e existir dicionario_hue, a ordem assumirá as chaves de dicionario_hue.
        Caso contrário, usará a ordem natural (unique()) dos dados.

    titulo : str, opcional
        Título do gráfico.

    rotacao_xticks : float, opcional
        Ângulo de rotação dos rótulos no eixo X (ticks).

    rotacao_xticks_barra : float, opcional
        Ângulo de rotação do texto que fica no topo de cada barra,
        indicando o valor de contagem.

    tamanho_texto : int, opcional
        Tamanho da fonte para o texto no topo das barras.

    paleta_cores : str ou list, opcional
        Paleta de cores Seaborn (ex.: 'Set2', 'viridis' ou uma lista de cores).
    """

    # Determinar a ordem do eixo X
    if ordem_x is None:
        if dicionario_x is not None:
            ordem_x = list(dicionario_x.keys())
        else:
            ordem_x = df[coluna_x].dropna().unique().tolist()

    # Determinar a ordem do hue
    if ordem_hue is None:
        if dicionario_hue is not None:
            ordem_hue = list(dicionario_hue.keys())
        else:
            ordem_hue = df[coluna_hue].dropna().unique().tolist()

    # Configurar tamanho da figura e DPI
    plt.figure(figsize=(10, 8), dpi=80)

    # Criar o gráfico de contagem
    ax = sns.countplot(
        data=df,
        x=coluna_x,
        hue=coluna_hue,
        palette=paleta_cores,
        order=ordem_x,
        hue_order=ordem_hue
    )

    # Adicionar valores no topo de cada barra
    for barra in ax.patches:
        altura = barra.get_height()
        if altura > 0:
            ax.text(
                barra.get_x() + barra.get_width() / 2.0,
                altura + 1,  # deslocamento fixo
                int(altura),
                ha="center",
                rotation=rotacao_xticks_barra,
                size=tamanho_texto
            )

    # Ajustar rótulos (labels) do eixo X com base no dicionario_x
    if dicionario_x:
        novos_rotulos_x = [dicionario_x.get(cat, str(cat)) for cat in ordem_x]
        ax.set_xticklabels(novos_rotulos_x)

    # Ajustar a legenda (hue) com base no dicionario_hue
    if dicionario_hue:
        # Obter 'handles' e 'labels' da legenda criada pelo Seaborn
        handles, labels = ax.get_legend_handles_labels()
        novos_rotulos_hue = []
        for rotulo in labels:
            try:
                # Tenta converter o rótulo para inteiro
                rotulo_int = int(rotulo)
                novo_rotulo = dicionario_hue.get(rotulo_int, rotulo)
            except ValueError:
                # Se não for possível converter, pode ser string
                novo_rotulo = dicionario_hue.get(rotulo, rotulo)
            novos_rotulos_hue.append(novo_rotulo)

        ax.legend(handles, novos_rotulos_hue,
                  title=coluna_hue.replace('_', ' ').title(),
                  loc='upper right')
    else:
        ax.legend(title=coluna_hue.replace('_', ' ').title(), loc='upper right')

    # Título, labels e ajustes finais
    plt.title(titulo, pad=20)
    plt.xlabel(coluna_x.replace('_', ' ').title())
    plt.ylabel('Quantidade')
    plt.xticks(rotation=rotacao_xticks)

    # Ajustar layout para não cortar nada
    plt.tight_layout()

    # Exibir gráfico
    plt.show()
```

```python
def plotar_histograma_porcentagem_por_hue(
    df,
    coluna_x,
    coluna_hue,
    dicionario_x=None,
    dicionario_hue=None,
    ordem_x=None,
    ordem_hue=None,
    titulo='',
    rotacao_xticks=0,
    rotacao_xticks_barra=0,
    tamanho_texto=12,
    paleta_cores='Set2'
):
    """
    Plota barras lado a lado (por hue), mostrando a porcentagem
    dentro de cada grupo do hue.

    Exemplo: Se 'coluna_hue' for "Sexo" (F/M), cada barra mostrará
    a porcentagem da contagem daquela faixa etária em relação ao
    total daquele hue (ou seja, total de F ou total de M).
    """
    # Montar um crosstab com contagens absolutas
    crosstab_absoluto = pd.crosstab(
        index=df[coluna_x],
        columns=df[coluna_hue]
    )
    # Convertendo para % (cada coluna somando 100%)
    crosstab_porcentagem = crosstab_absoluto.apply(lambda x: x / x.sum() * 100, axis=0)

    # Reformatar para formato "longo" (melt) que o Seaborn gosta
    df_plot = crosstab_porcentagem.reset_index().melt(
        id_vars=coluna_x,
        var_name=coluna_hue,
        value_name='porcentagem'
    )

    # definir uma ordem específica para x
    if ordem_x is None:
        if dicionario_x:
            ordem_x = list(dicionario_x.keys())
        else:
            ordem_x = sorted(crosstab_absoluto.index.unique())

    # definir uma ordem específica para hue
    if ordem_hue is None:
        if dicionario_hue:
            ordem_hue = list(dicionario_hue.keys())
        else:
            ordem_hue = sorted(crosstab_absoluto.columns.unique())

    plt.figure(figsize=(11, 8), dpi=80)
    ax = sns.barplot(
        data=df_plot,
        x=coluna_x,
        y='porcentagem',
        hue=coluna_hue,
        palette=paleta_cores,
        order=ordem_x,
        hue_order=ordem_hue
    )

    # Adiciona o valor % no topo de cada barra
    num_hue = len(ordem_hue)
    for i, barra in enumerate(ax.patches):
        altura = barra.get_height()
        x_pos = barra.get_x() + barra.get_width() / 2.0
        texto = f"{altura:.1f}%"
        ax.text(
            x_pos,
            altura + 0.5,
            texto,
            ha="center",
            rotation=rotacao_xticks_barra,
            size=tamanho_texto
        )

    # Ajustar rótulos do eixo X se existir dicionario_x
    if dicionario_x:
        novos_rotulos = [dicionario_x.get(xval, str(xval)) for xval in ordem_x]
        ax.set_xticklabels(novos_rotulos)

    # Ajustar a legenda se existir dicionario_hue
    if dicionario_hue:
        handles, labels = ax.get_legend_handles_labels()
        novos_rotulos = []
        for rotulo in labels:
            try:
                rotulo_int = int(rotulo)
                novo = dicionario_hue.get(rotulo_int, rotulo)
            except ValueError:
                novo = dicionario_hue.get(rotulo, rotulo)
            novos_rotulos.append(novo)
        # colocar a legenda para fora do grafico
        ax.legend(
            handles,
            novos_rotulos,
            title=coluna_hue,
            loc='upper left',
            bbox_to_anchor=(1, 1),
            borderaxespad=0
        )
    else:
        ax.legend(
            title=coluna_hue,
            loc='upper left',
            bbox_to_anchor=(1, 1),
            borderaxespad=0
        )

    plt.title(titulo, pad=15)
    plt.xlabel(coluna_x)
    plt.ylabel('Porcentagem dentro de cada grupo (%)')
    plt.xticks(rotation=rotacao_xticks)
    plt.tight_layout()
    plt.show()
```

```python
def plotar_histograma_proporcao(
    df,
    coluna_x,
    coluna_hue,
    dicionario_x=None,
    dicionario_hue=None,
    ordem_x=None,
    ordem_hue=None,
    titulo='',
    rotacao_xticks=0,
    rotacao_xticks_barra=0,
    tamanho_texto=12,
    paleta_cores='Set2'
):
    # usar Seaborn 0.12+ e stat='probability'
    if ordem_x is None:
        if dicionario_x:
            ordem_x = list(dicionario_x.keys())
        else:
            ordem_x = df[coluna_x].dropna().unique()

    if ordem_hue is None:
        if dicionario_hue:
            ordem_hue = list(dicionario_hue.keys())
        else:
            ordem_hue = df[coluna_hue].dropna().unique()

    plt.figure(figsize=(11, 8), dpi=80)
    ax = sns.countplot(
        data=df,
        x=coluna_x,
        hue=coluna_hue,
        palette=paleta_cores,
        order=ordem_x,
        hue_order=ordem_hue,
        stat='probability'  # faz as barras representarem proporção (0 a 1)
    )

    # Adiciona porcentagem no topo
    for barra in ax.patches:
        altura = barra.get_height()
        if altura > 0:
            ax.text(
                barra.get_x() + barra.get_width() / 2.0,
                altura + 0.001,
                f"{altura * 100:.1f}%",
                ha="center",
                rotation=rotacao_xticks_barra,
                size=tamanho_texto
            )

    # Ajustar rótulos do eixo X, se houver dicionário
    if dicionario_x:
        ax.set_xticklabels([dicionario_x.get(cat, str(cat)) for cat in ordem_x])

    # Ajustar a legenda (hue) fora do gráfico
    handles, labels = ax.get_legend_handles_labels()

    if dicionario_hue:
        novos_labels = []
        for lbl in labels:
            try:
                lbl_int = int(lbl)
                novo_label = dicionario_hue.get(lbl_int, lbl)
            except ValueError:
                novo_label = dicionario_hue.get(lbl, lbl)
            novos_labels.append(novo_label)
    else:
        novos_labels = labels

    # 
    ax.legend(
        handles,
        novos_labels,
        title=coluna_hue,
        loc='upper left',      # posição relativa
        bbox_to_anchor=(1, 1), # desloca a legenda para fora (lado direito)
        borderaxespad=0
    )

    plt.title(titulo, pad=15)
    plt.xlabel(coluna_x)
    plt.ylabel('Proporção')
    plt.xticks(rotation=rotacao_xticks)
    plt.tight_layout()
    plt.show()
```
