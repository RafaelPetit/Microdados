```python
"""
Este programa fara analise inicial dos dados do ENEM 2023

Este programa:
- abrirá o arquivo de dados do ENEM 2023
- fará uma análise inicial dos dados quanto a:
    - quantidade de linhas
    - quantidade de colunas
    - tipos de dados
- valores faltantes serão excluídos
- valores do tipo inteiro de 64 bits serão convertidos para inteiros de 8 bits
- valores do tipo float de 64 bits serão convertidos para inteiros de 16 bits

Como resultado o tamanho do arquivo foi reduzido de 251 MB para 159 MB

"""
```

```python
# Importando bibliotecas
import pandas as pd

# Abrindo o arquivo de dados
df = pd.read_csv('enem_2023_100k.csv', sep=';')

# Verificando se o arquivo foi aberto corretamente
print(df.head(), '\n')

# Verificando a quantidade de linhas e colunas
print(df.shape, '\n')
```

```python
# Configurando pandas para mostrar todas as colunas
pd.set_option('display.max_columns', 76)

# Verificando os tipos de dados de todas as colunas
print(df.dtypes.to_string(), '\n')
```

```python
# imprimindo o tamanho da memória ocupada pelo dataframe df em Mb com duas casas decimais
print(f'Memória ocupada pelo dataframe df: {df.memory_usage(deep=True).sum() / 1024 ** 2:.2f} MB\n')
```

```python
# Excluindo as colunas Q001 a Q025
# df.drop(columns=['Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006', 'Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012', 'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020', 'Q021', 'Q022', 'Q023', 'Q024', 'Q025'], axis = 1,inplace=True)

# Excluindo as colunas: NU_INSCRICAO, CO_MUNICIPIO_ESC, NU_ANO
df.drop(columns=['NU_INSCRICAO', 'CO_MUNICIPIO_ESC', 'CO_MUNICIPIO_PROVA', 'NU_ANO'], axis = 1, inplace=True)
```

```python
# imprimindo o tamanho da memória ocupada pelo dataframe df em Mb com duas casas decimais
print(f'Memória ocupada pelo dataframe df: {df.memory_usage(deep=True).sum() / 1024 ** 2:.2f} MB\n')
```

```python
# Verificando os tipos de dados de todas as colunas
print(df.dtypes.to_string(), '\n')
```

```python
# Verificando os valores faltantes
print('Valores faltantes por coluna: \n')
print(df.isnull().sum(), '\n')
```

```python
# Imprimindo os valores unicos de 'TP_ENSINO'
print(df['TP_ENSINO'].unique(), '\n')

# Substiotuindo os valores faltantes por 0 na coluna 'TP_ENSINO'
df['TP_ENSINO'] = df['TP_ENSINO'].fillna(0)

# Verificando os valores faltantes na coluna 'TP_ENSINO'
print(df['TP_ENSINO'].isnull().sum(), '\n')

# Imprimindo os valores unicos de 'TP_ENSINO'
print(df['TP_ENSINO'].unique(), '\n')

# Convertendo os valores float64 para int8
df['TP_ENSINO'] = df['TP_ENSINO'].astype('int8')

```

```python
# Imprimindo os valores unicos de 'TP_DEPENDENCIA_ADM_ESC'
print(df['TP_DEPENDENCIA_ADM_ESC'].unique(), '\n')

# Substituindo os valores faltantes por 0 na coluna 'TP_DEPENDENCIA_ADM_ESC'
df['TP_DEPENDENCIA_ADM_ESC'] = df['TP_DEPENDENCIA_ADM_ESC'].fillna(0)

# Verificando os valores faltantes na coluna 'TP_DEPENDENCIA_ADM_ESC'
print(df['TP_DEPENDENCIA_ADM_ESC'].isnull().sum(), '\n')

# Imprimindo os valores unicos de 'TP_DEPENDENCIA_ADM_ESC'
print(df['TP_DEPENDENCIA_ADM_ESC'].unique(), '\n')

# Convertendo os valores float64 para int8
df['TP_DEPENDENCIA_ADM_ESC'] = df['TP_DEPENDENCIA_ADM_ESC'].astype('int8')
```

```python
# Imprimindo os valores unicos de 'TP_LOCALIZACAO_ESC'
print(df['TP_LOCALIZACAO_ESC'].unique(), '\n')

# Substituindo os valores faltantes por 0 na coluna 'TP_LOCALIZACAO_ESC'
df['TP_LOCALIZACAO_ESC'] = df['TP_LOCALIZACAO_ESC'].fillna(0)

# Verificando os valores faltantes na coluna 'TP_LOCALIZACAO_ESC'
print(df['TP_LOCALIZACAO_ESC'].isnull().sum(), '\n')

# Imprimindo os valores unicos de 'TP_LOCALIZACAO_ESC'
print(df['TP_LOCALIZACAO_ESC'].unique(), '\n')

# Convertendo os valores float64 para int8
df['TP_LOCALIZACAO_ESC'] = df['TP_LOCALIZACAO_ESC'].astype('int8')
```

```python
# Imprimindo os valores unicos de 'SG_UF_ESC'
print(df['SG_UF_ESC'].unique(), '\n')

# Substituindo os valores faltantes por 'NI' na coluna 'SG_UF_PROVA'
df['SG_UF_ESC'] = df['SG_UF_ESC'].fillna('NI')

# Verificando os valores faltantes na coluna 'SG_UF_PROVA'
print(df['SG_UF_ESC'].isnull().sum(), '\n')

# Imprimindo os valores unicos de 'SG_UF_ESC'
print(df['SG_UF_ESC'].unique(), '\n')


```

```python
# Armazenando o nome das colunas com valores faltantes
colunas_faltantes = df.columns[df.isnull().sum() > 0]

# Excluindo as colunas com valores faltantes
df.drop(columns=colunas_faltantes, axis = 1, inplace=True)
```

```python
# Verificando os tipos de dados de todas as colunas
print(df.dtypes.to_string(), '\n')

# Verificando os valores únicos
print('Valores únicos por coluna: \n')
print(df.nunique(), '\n')
```

```python
# Verificando os valores máximos e mínimos
# Selecionando as colunas com valores numéricos inteiro  
numeric_int_df = df.select_dtypes(include=['int64'])
```

```python
# Verificando os valores máximos e mínimos das colunas numéricas inteiras
print('Valores máximos das colunas numéricas inteiras: \n')
print(numeric_int_df.max(), '\n')
print('Valores mínimos das colunas numéricas inteiras: \n')
print(numeric_int_df.min(), '\n')
```

```python
# Armazenando as colunas numéricas inteiras em uma lista
numeric_int_columns = numeric_int_df.columns.tolist()

# Convertendo as colunas numéricas int64 para inteiros sem sinal de 8 bits (uint8)
df[numeric_int_columns] = df[numeric_int_columns].astype('uint8')

# Verificando os tipos de dados de todas as colunas
print(df.dtypes.to_string(), '\n')
```
